```python
#creating a list and appending all the currencies into that list
curr=[]
for i in results:
    #print(i)
    if ('currencies' not in i.keys()):
        continue
    temp=i['currencies'][0]['code']
    if(temp not in curr):
        curr.append(temp)
```

```python
#creating a dictionary to have all data for each currency 
dict={}
for i in curr:
    dict[i]={'important': 0, 'toxic': 0, 'liked':0,'disliked':0,'positive':0, 'negative':0,'positive':0,'lol':0, 'saved':0, 'comments':0}
    print(dict)
```

```python
# Top words will be those with a frequency between 10 ans 30 (rule of thumb)
top_words = new[
    (new["Frequency"] >= 1) & (new["Frequency"] <= 50)
]

top_words.head(10)
```

```python
# Create a string list of terms to generate the word cloud
terms_list = str(top_words['Word'].tolist())

# Create the word cloud
wordcloud = WordCloud(width = 1000, height = 500,colormap="RdYlBu").generate(terms_list)
plt.axis("off")
fontdict = {"fontsize": 20, "fontweight": "bold"}
plt.title("Currency News Word Cloud", fontdict=fontdict)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud)
```
